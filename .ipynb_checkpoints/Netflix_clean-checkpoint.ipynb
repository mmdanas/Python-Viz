```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import calendar
from datetime import date
```

```python
df= pd.read_csv('Viewing activity.csv')
df.head()
```

```python
#splitting timestamp to date and time
#converting to timestamp format
df['Timestamp'] = pd.to_datetime(df['Timestamp'])
df['Day'] = df['Timestamp'].dt.weekday_name
df['Date']=pd.to_datetime(df['Timestamp']).dt.date
df['Time']=pd.to_datetime(df['Timestamp']).dt.time
```

```python
df.head()
```

```python
#bucketing based on timestamp
conditions = [
    (df['Time']> '08:00:00.000000') & (df['Time'] < '15:00:00.000000'),
    (df['Time']> '15:01:00.000000') & (df['Time'] < '20:00:00.000000'),
    (df['Time']> '20:01:00.000000') & (df['Time'] < '23:59:00.000000'),
    (df['Time']> '00:00:00.000000') & (df['Time'] < '07:59:00.000000')
]
values = ['Day Time','Evening','Night','Dawn']
df['WatchTime'] = np.select(conditions,values)
```

```python
df.drop(columns=['WatchTime',])
```
